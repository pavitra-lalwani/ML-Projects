```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pytrends.request import TrendReq
```

```python
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["jssaten"])
data = trends.interest_by_region()
data = data.sort_values(by = "jssaten", ascending=False)
```

```python
data = data.head(10)
```

```python
data = TrendReq(hl = "en-US",tz = 360)
data.build_payload(kw_list = ["jssaten"])
data = data.interest_over_time()
data.reset_index()
plt.figure(figsize = (20,15))
plt.plot(data["jssaten"])

plt.xlabel("YEAR")
plt.ylabel("Search Trends Count")
```
