```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import yfinance as yf
import datetime
from datetime import date,timedelta
import plotly.graph_objects as go
```

```python
today = date.today()
```

```python
today
```

```python
d1 = today.strftime("%Y-%m-%d")
end_dt = d1
d2 = date.today() - timedelta(days=1000)
d2 = d2.strftime("%Y-%m-%d")
start_dt = d2
eth = yf.download("ETH-INR",start=start_dt,end=end_dt,progress=False)
```

```python
eth
```

```python
eth["Date"] = eth.index
```

```python
eth
```

```python
eth = eth[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
```

```python
eth
```

```python
eth.reset_index(drop=True, inplace=True)
```

```python
eth
```

```python
eth.shape
```

```python
figure = go.Figure(data = go.Candlestick(x = eth["Date"],open = eth["Open"],high =eth["High"],low = eth["Low"],close = eth["Close"]))
figure.update_layout(title = "Ethereum Price Prediction")
figure.show()
```

```python
sns.heatmap(eth.corr(),cmap = "viridis")
```

```python
from autots import AutoTS
```

```python
fit_model = AutoTS(forecast_length=10,frequency="infer",ensemble="simple")
fit_model = fit_model.fit(eth, date_col='Date', value_col='Close', id_col=None)
```

```python
prediction = fit_model.predict()
```

```python
forecast = prediction.forecast
```
