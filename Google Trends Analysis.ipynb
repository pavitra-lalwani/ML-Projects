```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pytrends.request import  TrendReq
```

```python
trends = TrendReq()
```

```python
trends.build_payload(kw_list = ["Machine Learning"])
data = trends.interest_by_region()
data = data.sort_values(by = "Machine Learning", ascending = False)
```

```python
data = data.head(10)
```

```python
data.reset_index().plot(x = "geoName", y = "Machine Learning",kind = "bar",figsize = (20,15))
```

```python
data = TrendReq(hl='en-US',tz = 360)
data.build_payload(kw_list = ["Machine Learning"])
data = data.interest_over_time()
plt.figure(figsize = (20,15))
plt.plot(data["Machine Learning"])
plt.xlabel("YEAR")
plt.ylabel("Search Trends Count")
```
